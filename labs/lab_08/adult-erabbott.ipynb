```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.svm import LinearSVC
from sklearn.grid_search import GridSearchCV
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn import preprocessing 

# Enable inline Plots
%matplotlib inline
```

```python
cols = ['age',
        'workclass',
        'fnlwgt',
        'education',
        'education-num',
        'marital-status',
        'occupation',
        'relationship',
        'race',
        'sex',
        'capital-gain',
        'capital-loss,',
        'hours-per-week',
        'native-country',
        'income']
adult_df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',
                       header=0,names=cols,index_col=False)
```

```python
# remove ID number 'fnwlgig' column and the edu column bc the number corresponds

a2_df = adult_df.drop(['fnlwgt', 'education'], axis = 1)
```

```python
a2_df.replace(' ?', np.nan, inplace = True) ## goddamn space
```

```python
a2_df.income.replace({' <=50K':0, ' >50K':1 }, inplace = True)
```

```python
a2_df.sex.replace({' Male':0, ' Female':1 }, inplace = True)
```

```python
a2_df.isnull().sum()
```

```python
a2_df[a2_df.occupation.isnull()]
```

```python
mask = a2_df.occupation.isnull()
a3_df = a2_df[-mask]
a3_df.isnull().sum()
```

```python
mask = a3_df['native-country'].isnull()
a4_df = a3_df[-mask]
a4_df.isnull().sum()
```

```python
df = pd.get_dummies(a4_df)
```

```python
df.info()
```

```python
target = np.array(df['income'])
target.shape
```

```python
features = df.drop('income', axis = 1)
#scale post dummies and run stuff
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = .3, random_state = 12)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
y_pred = lr.predict(X_test)
confusion_matrix(y_test, y_pred)
```
