### Merge Electoral votes with pres_2004 to see who won and by how many electoral votes

* Merge vs Join

```python
pres_2004 = pres_2004.merge(electoral_votes) 
# be sure to assign the merge back to a variable (otherwise it'll just print)
```

```python
# Award the electoral Votes to the candidate that had the most votes in the stae
# How much did Kerry and Bush get?

#for row in pres_2004:
#    if Bush > Kerry:
#        Bush_elec += electoral votes
#    elif Kerry > Bush:
#        Kerry_elec += electoral votes
#    else:
#        Other_elec += electoral votes

#looping is inefficient so don't do that
# you want to compare which is larger and store it in another column

#create two new columns
pres_2004['Bush_votes'] = 0
pres_2004['Kerry_votes'] = 0
#mask is our 'filter'. Gives a boolean for each row so we know who won each section
bush_mask = pres_2004['Bush'] > pres_2004['Kerry']

pres_2004.ix[bush_mask, 'Bush_votes'] = pres_2004.ix[bush_mask, 'Electoral Votes']
pres_2004.loc[-bush_mask, 'Kerry_votes'] = pres_2004.loc[-bush_mask, 'Electoral Votes']
```

```python
pres_2004['Bush_votes'] = (pres_2004['Bush'] > pres_2004['Kerry']) * pres_2004['Electoral Votes']
pres_2004['Kerry_votes'] = (pres_2004['Bush'] < pres_2004['Kerry']) * pres_2004['Electoral Votes']
```

```python
print 'Kerry', pres_2004.Kerry_votes.sum()
print 'Bush', pres_2004.Bush_votes.sum()
```

```python
# Now add a NEW Column and set it to 2004.  We are going to merge files later
pres_2004['year'] = 2004
```
